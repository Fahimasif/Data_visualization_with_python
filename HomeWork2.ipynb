```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("ford.csv")
print(type(df))
display(df.head())
display(df.tail())
```

```python
df["transmission"].value_counts().plot( kind= "pie", autopct= "%1.1f%%", startangle= 180)
plt.title("Transmissions of ford car")
plt.ylabel("")
plt.show()
```

Find percentage of unique values of transmission 

```python
print(df["fuelType"].value_counts())
print(type(df["fuelType"].value_counts()))
df_fuelType = pd.DataFrame(df["fuelType"].value_counts())
display(df_fuelType.head())
print(df_fuelType.index)
print(df_fuelType.columns)
```

```python
df_fuelType = pd.DataFrame(df["fuelType"].value_counts())
df_fuelType = df_fuelType.reset_index()
df_fuelType = df_fuelType.rename(columns={"index":"fuelType", "fuelType":"no_of_cars"})

df_fuelType["% of cars"] = (df_fuelType["no_of_cars"]/df.shape[0])*100

df_fuelType = df_fuelType.round(2)
display(df_fuelType)
```

```python
sns.barplot(x="fuelType", 
            y="% of cars", 
            data=df_fuelType, 
            color="red",
            alpha=0.8)
plt.xlabel("Types of fules")
plt.ylabel("% of cars")
plt.title("percent of cars present in each fuelTypes")
plt.yticks(np.arange(0,101,10))
plt.show()
```
